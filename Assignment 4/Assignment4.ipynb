```python
"""Write a program to do following:
We have given a collection of 8 points. P1=[0.1,0.6] P2=[0.15,0.71] P3=[0.08,0.9] P4=[0.16,
0.85] P5=[0.2,0.3] P6=[0.25,0.5] P7=[0.24,0.1] P8=[0.3,0.2]. Perform the k-mean clustering
with initial centroids as m1=P1=Cluster#1=C1 and m2=P8=cluster#2=C2."""
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
points = [[0.1, 0.6], [0.15, 0.71], [0.08, 0.9], [0.16, 0.85],
          [0.2, 0.3], [0.25, 0.5], [0.24, 0.1], [0.3, 0.2]]


initial_centroids = [[0.1, 0.6], [0.3, 0.2]]
```

```python
kmeans = KMeans(n_clusters=2, init=initial_centroids, n_init=1).fit(points)

centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
for i, centroid in enumerate(centroids):
    points_in_cluster = [points[j] for j in range(len(points)) if labels[j] == i]
    print(f"Cluster {i+1}: Centroid {centroid} - Points {points_in_cluster}")
```

```python
cluster1_points = [points[i] for i in range(len(points)) if labels[i] == 0]
cluster2_points = [points[i] for i in range(len(points)) if labels[i] == 1]

plt.scatter([point[0] for point in cluster1_points], [point[1] for point in cluster1_points], color='red', label='Cluster 1')
plt.scatter([point[0] for point in cluster2_points], [point[1] for point in cluster2_points], color='blue', label='Cluster 2')
plt.scatter([centroid[0] for centroid in centroids], [centroid[1] for centroid in centroids], color='black', marker='x', label='Centroids')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('K-Means Clustering')
plt.legend()
plt.grid(True)
plt.show()
```
