```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("car_crashes.csv")
```

```python
df = pd.DataFrame(data)
```

```python
plt.scatter(df['speeding'], df['not_distracted'])
plt.xlabel('Speeding')
plt.ylabel('Not distracted')
plt.show()
```

```python
plt.bar(df['speeding'], df['alcohol'], width = 0.1)
plt.xlabel('Speeding')
plt.ylabel('Alcohol')
plt.show()
```

```python
df.boxplot(column = 'ins_premium')
```

```python
plt.hist(df['no_previous'], rwidth = 0.5)
plt.xlabel('No Previous')
plt.ylabel('Frequency')
plt.show()
```

```python
no_prev = df['no_previous'].sort_values()
plt.plot(no_prev, df['ins_premium'])
plt.xlabel('No Previous')
plt.ylabel('Insurance Premium')
plt.show()
```

```python
sns.violinplot(df['speeding'])
plt.show()
```

```python
sns.heatmap(data.corr(), cmap = "YlGnBu")
plt.show( )
```

```python
sns.pairplot(df)
plt.show()
```
